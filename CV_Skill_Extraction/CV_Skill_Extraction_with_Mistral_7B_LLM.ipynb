**Инициализация модели**

```python
import torch
from peft import AutoPeftModelForCausalLM  # Импорт класса AutoPeftModelForCausalLM из библиотеки peft
from transformers import AutoTokenizer, AutoModel  # Импорт классов AutoTokenizer и AutoModel из библиотеки transformers
import torch.nn.functional as F
from langchain.prompts import PromptTemplate

adapt_model_name = "IlyaGusev/saiga_mistral_7b_lora"
base_model_name = "Open-Orca/Mistral-7B-OpenOrca"

tokenizer = AutoTokenizer.from_pretrained(
              base_model_name,
              trust_remote_code=True)

tokenizer.pad_token = tokenizer.eos_token
device_map = {"": 0}
```

```python
import torch

if torch.cuda.is_available():
    device = torch.device("cuda")
    print("GPU доступен")
else:
    device = torch.device("cpu")
    print("GPU не доступен, используется CPU")
```

```python
model = AutoPeftModelForCausalLM.from_pretrained(
              adapt_model_name,
              device_map=device_map,
              torch_dtype=torch.bfloat16)
```

```python
import pandas as pd
```

```python
# Чтение текстового файла с помощью Pandas
df_dataset = pd.read_csv('/content/combined_fields.txt')
```

```python
# Создание промпта
question = "Выделить 7 или больше если есть ключевых навыков из текста. Навыки перечислить через запятую. Кроме навыков ничего не пишем. Не повторять ответ \nbot: Вот ответ на ваш вопрос длиной не более 20 слов:"
```

```python
# Добавление промпта к каждой строке и присвоение переменной
combined_with_prompt = df_dataset['Combined_fields'].apply(lambda x: f"{x} {question}")

# Вывод просуммированной строки для каждой строки
for summarized_line in combined_with_prompt[0:2]:
    print(summarized_line)
```

```python
# функция получения списка навыков
def get_answer(info_prompt):

    prompt = info_prompt.format(question=question)
    inputs = tokenizer(prompt, return_tensors="pt").to("cuda")
    outputs = model.generate(input_ids=inputs["input_ids"].to("cuda"),
                            top_p=0.5,
                            temperature=0.3,
                            attention_mask=inputs["attention_mask"],
                            max_new_tokens=50,
                            pad_token_id=tokenizer.eos_token_id,
                            do_sample=True)

    output = tokenizer.decode(outputs[0], skip_special_tokens=True)

    parsed_answer = output.split("Вот ответ на ваш вопрос длиной не более 20 слов:")[1].strip()

    if "bot:" in parsed_answer:
        parsed_answer = parsed_answer.split("bot:")[0].strip()

    return parsed_answer
```

```python
out = get_answer(combined_with_prompt[1])
```

```python
import time
skills = []
# Создаем пустой список для хранения времени выполнения каждой итерации
execution_times = []

for i in range(len(combined_with_prompt)):
    start_time = time.time()  # Записываем текущее время до выполнения кода

    out = get_answer(combined_with_prompt[i])
    skills.append(out)

    end_time = time.time()  # Записываем текущее время после выполнения кода
    execution_time = end_time - start_time  # Вычисляем время выполнения участка кода
    execution_times.append(execution_time)  # Добавляем время выполнения в список

# Вывести время выполнения каждой итерации
print("Execution times for each iteration:")
for i, exe_time in enumerate(execution_times):
    print(f"Iteration {i+1}: {exe_time} seconds")
```

Обработка навыков

```python
import re

class SkillsDataProcessor:
    def __init__(self, skills):
        self.skills = skills
        self.processed_skills = []

    def process_data(self):
        # Обработка каждой строки и добавление в новый список
        for item in self.skills:
            # Удаление лишних символов
            clean_item = re.sub(r'-', ' ', item)
            # Удаление лишних символов
            clean_item = re.sub(r'[^\w\s,]', ' ', item)
            # Разделение строки по запятой
            words = clean_item.split(',')
            # Добавление отдельных слов в список
            self.processed_skills.append(words)

    def remove_leading_spaces_and_chars(self):
        for sublist in self.processed_skills:
            for idx, item in enumerate(sublist):
                if isinstance(item, str):
                    # Удаляем пробелы и дефисы с начала строки
                    stripped_item = item.lstrip().lstrip('-')
                    sublist[idx] = stripped_item
        return self.processed_skills

    def capitalize_all_words(self):
        for sublist in self.processed_skills:
            for idx, word in enumerate(sublist):
                sublist[idx] = word.upper() if type(word) is str else word
        return self.processed_skills

    def forword(self):
        self.process_data()
        self.remove_leading_spaces_and_chars()
        self.capitalize_all_words()
        return self.processed_skills
```

```python
# создание нового экземпляра класса SkillsDataProcessor и инициализацию его с аргументом skills
processor = SkillsDataProcessor(skills)
```

```python
skl = processor.forword()
```
